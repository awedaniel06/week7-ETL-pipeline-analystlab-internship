```python
# ==========================
# BASIC ETL PIPELINE TO EXTRACTS REAL-TIME WEATHER DATA FROM OPENWEATHER API
# ==========================

# Import required libraries
import requests
import pandas as pd

# OpenWeather API Key
API_KEY = "96c1d43c202fb3167adf882714a85d46"

# List of cities
cities = ["Lagos", "Abuja", "Ibadan"]

# OpenWeather API endpoint
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

# Create an empty list to store weather data
weather_data = []

# Loop through each city
for city in cities:

    # Parameters to send to the API
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"
    }

    # Send GET request to the API
    response = requests.get(BASE_URL, params=params)

    # Check if request was successful
    if response.status_code == 200:

        # Convert response to JSON
        data = response.json()

        # Extract required information
        weather_data.append({
            "City": data["name"],
            "Country": data["sys"]["country"],
            "Temperature (°C)": data["main"]["temp"],
            "Feels Like (°C)": data["main"]["feels_like"],
            "Humidity (%)": data["main"]["humidity"],
            "Pressure (hPa)": data["main"]["pressure"],
            "Weather Condition": data["weather"][0]["description"],
            "Wind Speed (m/s)": data["wind"]["speed"],
            "Date & Time": pd.to_datetime(data["dt"], unit="s")
        })

        print(f"Successfully extracted data for {city}")

    else:
        print(f"Failed to retrieve data for {city}. Status Code: {response.status_code}")

# Convert the extracted data into a DataFrame
df = pd.DataFrame(weather_data)

# Display the extracted data
print("\nExtracted Weather Data:")
print(df)

# Save the data to a CSV file
df.to_csv("weather_data.csv", index=False)

print("\nWeather data has been successfully saved as 'weather_data.csv'.")
```

```python
# ==========================
# BASIC ANALYSIS
# ==========================

import pandas as pd

pd.read_csv('weather_data.csv')
print(pd.read_csv('weather_data.csv'))
df= pd.read_csv('weather_data.csv')


# Compare temperatures across cities
print("\nTemperature Across Cities")
print(df[["City", "Temperature (°C)"]])

# Find the city with the highest humidity
highest_humidity = df.loc[df["Humidity (%)"].idxmax()]

print("\nCity with the Highest Humidity")
print(highest_humidity[["City", "Humidity (%)"]])

# Compare weather conditions
print("\nWeather Conditions")
print(df[["City", "Weather Condition"]])

# Find the hottest city
hottest_city = df.loc[df["Temperature (°C)"].idxmax()]

print("\nHottest City")
print(hottest_city[["City", "Temperature (°C)"]])

# Find the coldest city
coldest_city = df.loc[df["Temperature (°C)"].idxmin()]

print("\nColdest City")
print(coldest_city[["City", "Temperature (°C)"]])

# Calculate the average temperature
average_temp = df["Temperature (°C)"].mean()

print("\nAverage Temperature")
print(round(average_temp, 2), "°C")
```
